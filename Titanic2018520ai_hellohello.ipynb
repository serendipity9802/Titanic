# Age
* 1.存在缺省值，需要填充
* 2.数据不连续，需要分类

```python
a = sns.FacetGrid(train, hue = 'Survived',aspect = 6)
a.map(sns.kdeplot,'Age',shade = True )
a.set(xlim=(0,train['Age'].max()))
a.add_legend()
```

观察数据，可以看出age主要集中在中间段，生存下来的比例也趋势基本一致。
* 1.缺省值采用随机数填充；
* 2.划分14|20|30|40|50|60，

```python
full_data = [train,test]
for dataset in full_data:
    Age_avg = dataset["Age"].mean()
    Age_std = dataset["Age"].std()
    size_isnull =dataset["Age"].isnull().sum()
    Age_random_list = np.random.randint(Age_avg-Age_std,Age_avg+Age_std,size = size_isnull)
    #dataset.loc[dataset["Age"].isnull(),"Age"] =Age_random_list
    dataset['Age'][np.isnan(dataset['Age'])] = Age_random_list
    
    dataset.loc[dataset["Age"]<14,"Age"] = 0
    dataset.loc[(dataset["Age"]>=14) & (dataset["Age"]<20),"Age"] = 1
    dataset.loc[(dataset["Age"]>=20) & (dataset["Age"]<30),"Age"] = 2
    dataset.loc[(dataset["Age"]>=30) & (dataset["Age"]<40),"Age"] = 3
    dataset.loc[(dataset["Age"]>=40) & (dataset["Age"]<50),"Age"] = 4
    dataset.loc[(dataset["Age"]>=50) & (dataset["Age"]<60),"Age"] = 5
    dataset.loc[dataset["Age"]>=60,"Age"] = 6
train["Age"].value_counts()
```

```python
sns.countplot('Age',data = train)

a = sns.FacetGrid(train, hue = 'Survived',aspect = 6)
a.map(sns.kdeplot,'Age',shade = True )
a.set(xlim=(0,train['Age'].max()))
a.add_legend()
```

# Fare 票价
票价训练集仅缺少一个，采用均值填充

```python
for dataset in full_data:
    dataset['Fare']=dataset['Fare'].fillna(train['Fare'].median())
```

```python
train.info()
```

```python
b = sns.FacetGrid(train, hue = 'Survived',aspect = 6)
b.map(sns.kdeplot,'Fare',shade = True )
b.set(xlim=(0,train['Fare'].max()))
b.add_legend()
```

&emsp;观察数据，这样的数据观察不出什么特征，  
但是数据都主要集中在左侧，需要对数据进行归一化处理，  
利用log

```python
for dataset in full_data:
    dataset['Fare'] = dataset['Fare'].map(lambda x:np.log(x) if x>0 else 0 )
```

```python
b = sns.FacetGrid(train, hue = 'Survived',aspect = 6)
b.map(sns.kdeplot,'Fare',shade = True )
b.set(xlim=(0,train['Fare'].max()))
b.add_legend()
```

```python
for dataset in full_data:
    dataset.loc[dataset['Fare']<1.4,'Fare'] =0
    dataset.loc[(dataset["Fare"]>=1.4) &(dataset['Fare']<2.7),'Fare'] =1
    dataset.loc[dataset["Fare"]>=2.7,'Fare']=2
```

```python
sns.countplot('Fare',hue="Survived",data =train )
```

这样就能看出各类的票价情况下，生存的比例了  
当票价偏低时，生存几率小；  
当票价为中间时，生存几率提高；  
当票价大时，生存几率更大 

# Embarked 
训练数据有两个缺失值，需要填补

```python
train["Embarked"].value_counts()
```

* Embarked 中，“S”居多，采用多数进行填充  
* 其次，需要将其数值化，便于计算

```python
for dataset in full_data:
    dataset["Embarked"]= dataset["Embarked"].fillna("S")
    dataset["Embarked"] = dataset["Embarked"].map({"S":0,"C":1,"Q":2}).astype(int)
```

# Name 

```python
train["Name"].value_counts()
```

```python
import re
def get_title(name):
    title_search = re.search('([A-Za-z]+)\.',name)   
    if title_search:
        return title_search.group(1)
    else:
        return ""
for dataset in full_data:
    dataset['Title']=dataset["Name"].apply(get_title)

train["Title"].value_counts()
```

```python
fig,(ax1) =plt.subplots(1,figsize=(18,6))
sns.barplot(x="Title",y="Survived",data=train,ax=(ax1))
```

对出现比较多的前缀进行映射，"Mr":1,"Miss":2,"Mrs":3,"Master":4,"Dr":5 ,"Rev":6，其他为0

```python
title_dict ={"Mr":1,"Miss":2,"Mrs":3,"Master":4,"Dr":5 ,"Rev":6}
for dataset in full_data:
    dataset["Title"]=dataset["Title"].map(lambda x:title_dict[x] if x in title_dict.keys() else 0)
```

```python
train["Title"].value_counts()
```

# 特征选择
选择不同数量的特征，对于结果的影响  
从十个特征中选择["Pclass",'Title','Sex','Age','Parch','Fare','Embarked','Has_Cabin','SibSp','Alone']

```python
feature_list =["Pclass",'Title','Sex','Age','Parch','Fare','Embarked','Has_Cabin','SibSp','Alone']
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(train[feature_list].astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
#X_train = train[['Pclass']]
from sklearn.tree import DecisionTreeClassifier
def different_feature(choose_feature_list):
    X_train = train[choose_feature_list]
    Y_train = train['Survived']
    X_test = test[choose_feature_list]
    dtree = DecisionTreeClassifier()
    dtree.fit(X_train,Y_train)
    prediction = dtree.predict(X_test)

    dfPrediction = pd.DataFrame(data= prediction ,index = test.index.values,columns = ['Survived'])
    contentTestPredObject1 = dfPrediction.to_csv()
    #print(contentTestPredObject1)     

    #y_train = test['Survived']
    acc_decision_tree = round(dtree.score(X_train, Y_train) * 100, 2)
    print('choose_feature_list:%s'%str(choose_feature_list),'\nacc:%.2f'%acc_decision_tree)
    return choose_feature_list,acc_decision_tree
different_feature(choose_feature_list=["Pclass"])
```

```python
different_feature(choose_feature_list=["Pclass",'Title'])
```

```python
different_feature(choose_feature_list=["Pclass",'Title','Sex'])
```

```python
different_feature(choose_feature_list=["Pclass",'Title','Sex','Age'])
```

```python
All_feature=[]
All_acc =[]
for i in range(1,len(feature_list)+1):
    choose_feature,acc=different_feature(feature_list[0:i])
    All_feature.append(str(choose_feature))
    All_acc.append(acc)
```

```python
plt.bar(range(len(All_feature)),All_acc, align='center', alpha=0.5, color='r')
plt.xticks(range(len(All_feature)),All_feature,  rotation='vertical')
plt.ylabel('Accuracy')
plt.title('choose different fearue with decision_tree')
plt.show()
```

```python
from sklearn import ensemble
#rf = ensemble.RandomForestRegressor(n_estimators = 20)
def different_feature(choose_feature_list):
    X_train = train[choose_feature_list]
    Y_train = train['Survived']
    X_test = test[choose_feature_list]
    dtree = ensemble.RandomForestRegressor(n_estimators = 10)
    dtree.fit(X_train,Y_train)
    prediction = dtree.predict(X_test)

    dfPrediction = pd.DataFrame(data= prediction ,index = test.index.values,columns = ['Survived'])
    contentTestPredObject1 = dfPrediction.to_csv()
    #print(contentTestPredObject1)     

    #y_train = test['Survived']
    acc_decision_tree = round(dtree.score(X_train, Y_train) * 100, 2)
    print('choose_feature_list:%s'%str(choose_feature_list),'\nacc:%.2f'%acc_decision_tree)
    return choose_feature_list,acc_decision_tree

All_feature=[]
All_acc =[]
for i in range(1,len(feature_list)+1):
    choose_feature,acc=different_feature(feature_list[0:i])
    All_feature.append(str(choose_feature))
    All_acc.append(acc)
plt.bar(range(len(All_feature)),All_acc, align='center', alpha=0.5, color='r')
plt.xticks(range(len(All_feature)),All_feature,  rotation='vertical')
plt.ylabel('Accuracy')
plt.title('choose different fearue with decision_tree')
plt.show()
```
